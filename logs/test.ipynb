```python
import torch
import matplotlib.pyplot as plt
```

```python
H_ = torch.load("../error.pt")["H_2"]
```

```python
damping = torch.mean(torch.abs(torch.diag(H_)))*0.0
print(damping)

H = H_.clone().float()
indexs = torch.arange(H.shape[0])
H[indexs, indexs] += damping

plt.imshow(torch.linalg.cholesky(H).cpu().numpy())
```

```python
damping = torch.mean(torch.abs(torch.diag(H_)))*0.00001
print(damping)

H = H_.clone().float()
indexs = torch.arange(H.shape[0])
H[indexs, indexs] += damping

plt.imshow(torch.linalg.cholesky(H).cpu().numpy())
```
